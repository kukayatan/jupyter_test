```python
# this script helps to visualize the results of the given function in the graph

import matplotlib.pyplot as plt

def function1(*args):
    result = []
    for arg in args:
# defining the function which results has to be plotted in the graph
        result.append((-2 * arg ** 2) + 3)    
    return result

# defining the range of x values [start,end,steps]
init = []
x = range(-8,9,1)
for n in x:
    init.append(n) 

res = function1(*init)
```

```python
# plotting the calculated results of the given function
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.plot(init, res, linewidth=2,linestyle="--")
plt.title("Plotted Results")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
